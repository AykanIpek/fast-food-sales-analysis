```python
import pandas as pd
import numpy as np
```

```python
menu_items = pd.read_csv('/content/drive/MyDrive/Data Science Proje 2/Fast-Food Restaurant Chain/menu_items.csv')
menuitem = pd.read_csv('/content/drive/MyDrive/Data Science Proje 2/Fast-Food Restaurant Chain/menuitem.csv')
pos_ordersale = pd.read_csv('/content/drive/MyDrive/Data Science Proje 2/Fast-Food Restaurant Chain/pos_ordersale.csv')
store_restaurant = pd.read_csv('/content/drive/MyDrive/Data Science Proje 2/Fast-Food Restaurant Chain/store_restaurant.csv')
```

```python
merged_data = pd.merge(menu_items,menuitem,left_on=["PLU","MenuItemId"],right_on=["PLU","Id"],how="outer")
merged_data.head()
```

```python
pos_ordersale.head()
```

```python
pos_ordersale.info()
```

```python
merged_data = pd.merge(merged_data,pos_ordersale,on="MD5KEY_ORDERSALE",how="outer")
merged_data.head()
```

```python
store_restaurant.head()
```

```python
merged_data = pd.merge(merged_data,store_restaurant,left_on="StoreNumber_y",right_on="STORE_NUMBER",how="outer")
merged_data.head()
```

```python
missing_summary = pd.DataFrame({
    "NaN_count" : merged_data.isnull().sum(),
    "Filled_count" : merged_data.notnull().sum(),
    "Total" : len(merged_data)
})
print(missing_summary)
```
